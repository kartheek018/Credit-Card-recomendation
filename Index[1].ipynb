### Above 30 >=5 and above 50 <=3

```python
spendings_above[(spendings_above["Above_30"]>=5) & (spendings_above["Above_50"]<3)]  
```

```python
def usage_septile(row):
    if row["Above_50"]>=5:
        return 1
    elif row["Above_50"]>=3 and row["Above_50"]<5:
        return 2
    elif row["Above_30"]>=5 :
        return 3
    elif row["Above_30"]>=3 and row["Above_30"]<5:
        return 4
    elif row["Above_30"]>=1 and row["Above_30"]<=2:
        return 5
    else:
        return 6
    
```

```python
spendings_above["Usage_septile"]=spendings_above.apply(usage_septile,axis=1)
```
